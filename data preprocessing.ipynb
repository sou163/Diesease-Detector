```python
# setting up training and testing folder paths
train_path = "/content/drive/MyDrive/fruit-and-vegetable-image-recognition/training"
test_path = "/content/drive/MyDrive/fruit-and-vegetable-image-recognition/test"
```

```python
import pathlib
import random
import numpy as np
import pandas as pd
import matplotlib.image as mpimg
```

```python
data_dir = pathlib.Path(train_path) # turning training path into a Python path
class_list = sorted([item.name for item in data_dir.glob('*')])
class_names = np.array(class_list) # obtain list of class_names
print(class_names)
```

```python
# viewing a random image
def view_image(target_dir, target_class):
    # target directory
    target_folder = target_dir + "/" + target_class
    
    # random image path
    random_image = random.sample(os.listdir(target_folder), 50)

    # Read in the image and plot it using matplotlib
    img = mpimg.imread(target_folder + "/" + random_image[0])
    plt.imshow(img)
    plt.title(target_class)
    plt.axis("off");

    return img
```

```python
plt.figure(figsize = (15,10))

# Viewing a random image from all classes in the training dataset
for i in range(len(class_names)):
    plt.subplot(5, 8, i+1)
    img = view_image(target_dir = train_path, 
                            target_class = class_names[i])
plt.savefig("/content/drive/MyDrive/fruit-and-vegetable-image-recognition/random_image")
```

```python
# preparing training dataframe
train_val_data = {'FilePath' : [],
                  'FileName': [],
                  'Label': []}

for dirpath, dirs, files in os.walk(train_path):
    for file in files:
        train_val_data['FilePath'].append(dirpath)
        train_val_data['FileName'].append(file)
        train_val_data['Label'].append(file.split('_')[0])
```

```python
train_val_data_df = pd.DataFrame(train_val_data)
train_val_data_df.to_csv('/content/drive/MyDrive/fruit-and-vegetable-image-recognition/train_val.csv')
train_val_data_df.head()
```

```python
# preparing test dataframe
test_data = {'FilePath' : [],
             'FileName': []}

for dirpath, dirs, files in os.walk(test_path):
    for file in files:
        test_data['FilePath'].append(dirpath)
        test_data['FileName'].append(file)
```

```python
test_data_df = pd.DataFrame(test_data)
test_data_df.to_csv('/content/drive/MyDrive/fruit-and-vegetable-image-recognition/test.csv')
test_data_df.head()
```
